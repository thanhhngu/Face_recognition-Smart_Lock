```python
import face_recognition
import cv2
import pickle

# Load data from encodings
with open("encodings.pkl", "rb") as f:
    known_encodings, known_names = pickle.load(f)

#webcam
cap = cv2.VideoCapture(0)

replace_encodings = []
replace_names = []

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # convert rgb
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # face encodings
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)
    # find face in encodings.pkl
    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        matches = face_recognition.compare_faces(known_encodings, face_encoding)
        name = "Unknown"

        #get face
        face_distances = face_recognition.face_distance(known_encodings, face_encoding)
        if len(face_distances) > 0:
            best_match_index = face_distances.argmin()
            similarity = (1 - face_distances[best_match_index]) * 100
            
            #arr replace
            if similarity > 79:
                if len(replace_encodings) < 3:
                    replace_encodings.append(face_encoding)
                    replace_names.append(known_names[best_match_index])
                
                
        if matches[best_match_index]:
            name = f"{known_names[best_match_index]} ({similarity:.2f}%)"

        #return infor
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
        cv2.putText(frame, name, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        ####

        ####
    #end programme
    cv2.imshow("Face Recognition Realtime", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
#replace
for new_encoding, target_name in zip(replace_encodings, replace_names):
    indices = [i for i, name in enumerate(known_names) if name == target_name]
    if indices:
        replace_idx = random.sample(indices, len(replace_encodings))
        known_encodings[replace_idx] = new_encoding

with open("encodings.pkl", "wb") as f:
    pickle.dump((known_encodings, known_names), f)

cap.release()
cv2.destroyAllWindows()
```
