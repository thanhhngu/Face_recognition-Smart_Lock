```python
import cv2
import os
import time
import pickle
import face_recognition

label = input('nhap label: ')

#addres save jpg
save_dir = os.path.join("C:/Users/ADMIN/OneDrive/workSpace/face-recognition/data", label)
os.makedirs(save_dir, exist_ok=True)

#set cam
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 800)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 600)

#check webcam
while True:
    ret, frame = cap.read()
    if not ret:
        print("error.")
        break
    cv2.imshow("Webcam", frame)
    if cv2.waitKey(1) != -1:
        break

#delay time
target_frames = 50
delay = 6 / target_frames
i = 0

while i < target_frames:
    ret, frame = cap.read()
    if not ret:
        break
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    face_locations = face_recognition.face_locations(rgb_frame)

    #create rectangle
    for (top, right, bottom, left) in face_locations:
         cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)

    #
    cv2.imshow("Webcam", frame)

    #save file
    filename = os.path.join(save_dir, f"{label}_{i:03d}.jpg")
    for (top, right, bottom, left) in face_locations:
        face_image = frame[top:bottom, left:right]
        filename = os.path.join(save_dir, f"{label}_{i:03d}.jpg")
        cv2.imwrite(filename, face_image)
    print(f"save: {filename}")
    i += 1

    #break 
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    
    time.sleep(delay)


data_dir = "C:/Users/ADMIN/OneDrive/workSpace/face-recognition/data"
known_encodings = []
known_names = []

#overload file encodings
if os.path.exists("encodings.pkl"):
    with open("encodings.pkl", "rb") as f:
        old_encodings, old_names = pickle.load(f)
        known_encodings.extend(old_encodings)
        known_names.extend(old_names)

#data encryption

for name in os.listdir(data_dir):
    person_dir = os.path.join(data_dir, name)
    if not os.path.isdir(person_dir):
        continue

    for filename in os.listdir(person_dir):
        img_path = os.path.join(person_dir, filename)
        image = face_recognition.load_image_file(img_path)
        encodings = face_recognition.face_encodings(image)

        if encodings:
            known_encodings.append(encodings[0])
            known_names.append(name)
#save data encryption
with open("encodings.pkl", "wb") as f:
    pickle.dump((known_encodings, known_names), f)

print("Đã mã hóa và lưu dữ liệu khuôn mặt.")

#end cam
cap.release()
cv2.destroyAllWindows()
```
